```python

# 1
# Update Recorded Damages
def convert_damage(damages):
    conversion = {"M": 1000000,
                  "B": 1000000000}
    update_damages=[]
    for damage in damages:
        if damage =='Damages not recorded':
            update_damages.append(damage)
        if damage.find('M') != -1:
            num = float(damage[0:damage.find('M')]) * conversion["M"]
            update_damages.append(num)
        if damage.find('B') != -1:
            num = float(damage[0:damage.find('B')]) * conversion["B"]
            update_damages.append(num)
    return update_damages

# test function by updating damages
updated_damages = convert_damage(damages)

# 2 
# Create a Dictionary
def create_dictionary(names, months, years, max_sustained_winds, areas_affected, updated_damages, deaths):
    hurricanes = {}
    for i in range(len(names)):
        hurricanes[names[i]]= {'Name': names[i], 
                              'Month': months[i], 
                              'Year': years[i], 
                              'Max Sustained Wind': max_sustained_winds[i], 
                              'Areas Affected': areas_affected[i], 
                              'Damage': updated_damages[i], 'Deaths': deaths[i]}
    return hurricanes

hurricanes = (create_dictionary(names, months, years, max_sustained_winds, areas_affected, updated_damages, deaths))


# 3
# Organizing by Year
def create_dictionary_year(hurricanes):
    hurricanes_year={}
    for hurricane in hurricanes:
        year = hurricanes[hurricane]['Year']
        current_hurricane = hurricanes[hurricane]
        if year not in hurricanes_year:
            hurricanes_year[year]= [current_hurricane]
        else:
            hurricanes_year[year].append(current_hurricane)
    return hurricanes_year
        

# create a new dictionary of hurricanes with year and key
hurricanes_year=create_dictionary_year(hurricanes)


# 4
# Counting Damaged Areas

def count_affected_areas(hurricanes):
    affected_areas_count = {}
    for hurricane in hurricanes:
        for area in hurricanes[hurricane]['Areas Affected']:
            if area not in affected_areas_count:
                affected_areas_count[area] = 1
            else:
                affected_areas_count[area] +=1
    return affected_areas_count


# create dictionary of areas to store the number of hurricanes involved in
affected_areas_count = count_affected_areas(hurricanes)


# 5 
# Calculating Maximum Hurricane Count
def most_affected_area(affected_areas_count):
    max_area = ''
    max_area_count = 0
    for area in affected_areas_count:
        if affected_areas_count[area] >max_area_count:
            max_area = area
            max_area_count= affected_areas_count[area]
            
    return {max_area, max_area_count}


# find most frequently affected area and the number of hurricanes involved in
most_freq = most_affected_area(affected_areas_count)


# 6
# Calculating the Deadliest Hurricane
def deadliest_hurricane(hurricanes):
    max_mortality_name = ''
    max_deaths = 0
    for cane in hurricanes:
        n_deaths = hurricanes[cane]['Deaths']
        if n_deaths> max_deaths:
            max_deaths = n_deaths
            max_mortality_name = cane
    return {max_mortality_name,max_deaths}
    

# find highest mortality hurricane and the number of deaths
#print(deadliest_hurricane(hurricanes))

# 7
# Rating Hurricanes by Mortality

def hurricane_mortality_scale(hurricanes):
    mortality_scale = {0: 0,
                   1: 100,
                   2: 500,
                   3: 1000,
                   4: 10000}
    hurricanes_by_mortality = {0:[],1:[],2:[],3:[],4:[],5:[]}
    for cane in hurricanes:
        n_deaths = hurricanes[cane]['Deaths']
        if n_deaths == 0:
            hurricanes_by_mortality[0].append(hurricanes[cane])
        elif n_deaths> 0 and n_deaths<=100:
            hurricanes_by_mortality[1].append(hurricanes[cane])
        elif n_deaths> 100 and n_deaths<=500:
            hurricanes_by_mortality[2].append(hurricanes[cane])
        elif n_deaths> 500 and n_deaths<=1000:
            hurricanes_by_mortality[3].append(hurricanes[cane])
        elif n_deaths> 1000 and n_deaths<=10000:
            hurricanes_by_mortality[4].append(hurricanes[cane])
        else:
            hurricanes_by_mortality[5].append(hurricanes[cane])
    return hurricanes_by_mortality


# categorize hurricanes in new dictionary with mortality severity as key
hurricanes_by_mortality = (hurricane_mortality_scale(hurricanes))

#print(hurricanes_by_mortality[5])
# 8 Calculating Hurricane Maximum Damage
def hurricane_maximum_Damage(hurricanes):
    max_damage_cane = ''
    max_damage = 0
    for cane in hurricanes:
        n_damage = hurricanes[cane]['Damage']
        if n_damage == 'Damages not recorded':
            pass
        elif n_damage > max_damage:
            max_damage=n_damage
            max_damage_cane= cane
            
    return {max_damage_cane, max_damage}


# find highest damage inducing hurricane and its total cost
print(hurricane_maximum_Damage(hurricanes))


# 9
# Rating Hurricanes by Damage

def hurricane_damage_scale(hurricanes):
    damage_scale = {0: 0,
                1: 100000000,
                2: 1000000000,
                3: 10000000000,
                4: 50000000000}
    hurricanes_by_damage = {0:[],1:[],2:[],3:[],4:[],5:[]}
    for cane in hurricanes:
        n_damage = hurricanes[cane]['Damage']
        if n_damage== 'Damages not recorded':
            hurricanes_by_damage[0].append(hurricanes[cane])
        
        elif n_damage == damage_scale[0]:
            hurricanes_by_damage[0].append(hurricanes[cane])
        elif n_damage > damage_scale[0] and n_damage <= damage_scale[1]:
            hurricanes_by_damage[1].append(hurricanes[cane])
            
        elif n_damage > damage_scale[1] and n_damage <= damage_scale[2]:
            hurricanes_by_damage[2].append(hurricanes[cane])
            
        elif n_damage > damage_scale[2] and n_damage <= damage_scale[3]:
            hurricanes_by_damage[3].append(hurricanes[cane])
            
        elif n_damage > damage_scale[3] and n_damage <= damage_scale[4]:
            hurricanes_by_damage[4].append(hurricanes[cane])
        
        elif n_damage > damage_scale[4]:
            hurricanes_by_damage[5].append(hurricanes[cane])
    return hurricanes_by_damage
  
# categorize hurricanes in new dictionary with damage severity as key
print(hurricane_damage_scale(hurricanes)[5])
```
