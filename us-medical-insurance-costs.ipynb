```python
ages = []
sexes = []
bmis = []
num_children = []
smoker_statuses = []
regions = []
insurance_charges = []
```

```python
def load_data(lst, dataset, name):
    with open(dataset) as csv_info:
        csv_dictionary = list(csv.DictReader(csv_info))
        for i in csv_dictionary:
            lst.append(i[name])
    return lst
```

```python
load_data(ages, 'insurance.csv', 'age')
load_data(sexes, 'insurance.csv', 'sex')
load_data(bmis, 'insurance.csv', 'bmi')
load_data(num_children, 'insurance.csv', 'children')
load_data(smoker_statuses, 'insurance.csv', 'smoker')
load_data(regions, 'insurance.csv', 'region')
load_data(insurance_charges, 'insurance.csv', 'charges')
ages
```

```python
class Patients_info:
    def __init__(self, patients_ages, patients_sexes, patients_bmis, 
    patients_num_children,patients_smoker_statuses, patients_regions, patients_charges):
        self.patients_ages = patients_ages
        self.patients_sexes = patients_sexes
        self.patients_bmis = patients_bmis
        self.patients_num_children = patients_num_children
        self.patients_smoker_statuses = patients_smoker_statuses
        self.patients_regions = patients_regions
        self.patients_charges = patients_charges
    
    def average_age(self):
        total_age = 0
        ages= self.patients_ages
        for age in ages:
            total_age +=int(age)
        print ("Average Age: " + str(round(total_age/len(ages)))+ " years")

    def number_per_sex(self):
        male = 0
        female = 0
        sexes = self.patients_sexes
        for sex in sexes:
            if sex == 'male':
                male+=1
            else:
                female +=1
        print(f"Number of male: {male}")
        print(f"Number of female: {female}")
    
    def unique_regions(self):
        unique_regions= []
        regions = self.patients_regions
        for region in regions:
            if region not in unique_regions:
                unique_regions.append(region)
        print(unique_regions)
    
    def average_insurance(self):
        total_charges = 0
        charges = self.patients_charges
        for charge in charges:
            total_charges += float(charge)
        avg_charge= round(total_charges/len(charges) , 2)
        print("Average insurance: " + str(avg_charge) + " $")
    
    def create_dict(self):
        self.patients_dict = {}
        self.patients_dict["age"] = [int(age) for age in self.patients_ages]
        self.patients_dict["sex"] = self.patients_sexes
        self.patients_dict["bmi"] = [float(bmi) for bmi in self.patients_bmis]
        self.patients_dict["children"] =[int(num_children) for num_children in self.patients_num_children] 
        self.patients_dict["smoker"] = self.patients_smoker_statuses
        self.patients_dict["regions"] = self.patients_regions
        self.patients_dict["charges"] = [float(charge) for charge in self.patients_charges]
        return self.patients_dict
```

```python
p=Patients_info(ages, sexes, bmis, num_children, smoker_statuses, regions, insurance_charges)
p.average_age()
p.number_per_sex()
p.unique_regions()
p.average_insurance()
```
